```python
from __future__ import print_function
import numpy as np
np.random.seed(1337)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer

import numpy as np
from keras.models import Sequential
from keras.callbacks import ModelCheckpoint, CSVLogger
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
```

```python
dataset = np.loadtxt("./Dataset/numerical_data_processed_new.csv", delimiter=",")
X = dataset[:,0:8]
Y = dataset[:,8]

scaler = Normalizer().fit(X)
X = scaler.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0],X_train.shape[1],1))
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
```

```python
cnn = Sequential()
cnn.add(Conv1D(64, 3, padding="same", activation="relu", input_shape=(8, 1)))
cnn.add(MaxPooling1D(pool_size=2))
cnn.add(Flatten())
cnn.add(Dense(128, activation="relu"))
cnn.add(Dropout(0.5))
cnn.add(Dense(1, activation="sigmoid"))
```

```python
cnn.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])

# train
checkpointer = ModelCheckpoint(filepath="logs/cnn/checkpoint-{epoch:02d}.hdf5", verbose=1, save_best_only=True, monitor='val_accuracy', mode='max')
csv_logger = CSVLogger('logs/cnn/cnntrainanalysis1.csv', separator=',', append=False)
cnn.fit(X_train, y_train, epochs=50, verbose=1, validation_data=(X_test, y_test), callbacks=[checkpointer, csv_logger])
cnn.save("./Models/cnn_model.hdf5")

loss, accuracy = cnn.evaluate(X_test, y_test)
print("\nLoss: %.2f, Accuracy: %.2f%%" % (loss, accuracy * 100))
```
